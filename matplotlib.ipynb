```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=z, cmap='plasma')
fig.colorbar(sc, label='Z intensity')
ax.set_title("3D Scatter Plot",color='black')
ax.set_xlabel("X axis",color='black')
ax.set_ylabel("Y axis",color='black')
ax.set_zlabel("Z axis",color='black')
plt.show()
```

```python
x = np.linspace(-6, 6, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='coolwarm')
fig.colorbar(surf)
plt.title("3D Surface Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], 'r-')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    return line,

def update(frame):
    x_data.append(frame)
    y_data.append(frame**2)
    line.set_data(x_data, y_data)
    return line,

ani = animation.FuncAnimation(fig, update, frames=range(21), init_func=init, blit=True)
plt.title("Animated y = x²")
plt.close(fig) # Close the plot window to prevent it from displaying during saving
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], 'r-')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    return line,

def update(frame):
    x_data.append(frame)
    y_data.append(frame**2)
    line.set_data(x_data, y_data)
    return line,

# Recreate the animation object with blit=False for saving
ani_save = animation.FuncAnimation(fig, update, frames=range(21), init_func=init, blit=False)

# Save the animation
ani_save.save("animation.mp4", writer='ffmpeg', fps=1)

plt.close(fig) # Close the plot window after saving
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import numpy as np

x = np.linspace(0, 10, 1000)
a = 1
y = np.sin(2 * np.pi * 0.1 * x)


fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

line, = ax.plot(x, y)

# Slider axis: [left, bottom, width, height]
slider_ax = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(slider_ax, 'Amplitude', 0.1, 2.0, valinit=a)

def update(val):
    line.set_ydata(slider.val * np.sin(2 * np.pi * 0.1 * x))
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.title("Interactive Sine Wave")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/student_marks.csv")

# 1. Line Chart - Individual subject performance of a student (e.g., Alice)
subjects = ['Math', 'Science', 'English', 'History']
alice_scores = data[data['Name'] == 'Alice'][subjects].values[0]

# 2. Bar Chart - Average scores of each subject
average_scores = data[subjects].mean()

# 3. Pie Chart - Contribution of each subject to Charlie’s total score
charlie_scores = data[data['Name'] == 'Charlie'][subjects].values[0]

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Line Chart
axs[0, 0].plot(subjects, alice_scores, marker='o', color='blue')
axs[0, 0].set_title("Alice's Performance")
axs[0, 0].set_ylim(0, 100)

# Plot 2: Bar Chart
axs[0, 1].bar(subjects, average_scores, color='green')
axs[0, 1].set_title("Average Marks in Each Subject")
axs[0, 1].set_ylim(0, 100)

# Plot 3: Pie Chart
axs[1, 0].pie(charlie_scores, labels=subjects, autopct='%1.1f%%', startangle=90)
axs[1, 0].set_title("Charlie's Subject-wise Contribution")

# Plot 4: Scatter - Math vs Science for all students
axs[1, 1].scatter(data['Math'], data['Science'], color='purple')
axs[1, 1].set_title("Math vs Science")
axs[1, 1].set_xlabel("Math Marks")
axs[1, 1].set_ylabel("Science Marks")

plt.tight_layout()
plt.savefig("student_dashboard.png", dpi=300)
plt.show()
```
