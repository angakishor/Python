```python
import re
def extract_hashtags(text):
  regex = r"#(\w+)|@(\w+)"
  hashtag_list = re.findall(regex, text)
  print("The hashtags in \"" + text + "\" are:")
  for hashtag in hashtag_list:
    print(hashtag)
if __name__ == "__main__":
  extract_hashtags(str(spl.Caption))
```
